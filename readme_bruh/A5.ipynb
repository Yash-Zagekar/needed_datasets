```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.bookswagon.com/promo-best-seller/minimum-40-off/20E06FE2A5BC")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
def remove_whitespace(text):
    """Remove all types of whitespace from the given text."""
    return ''.join(text.split())

# Example usage
cleaned_text = remove_whitespace(page_data)
print(cleaned_text)  # Output: "Hello,howareyou?Thisisatest."
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
        """
        ### SCRAPED TEXT FROM WEBSITE:
        {page_data}
        ### INSTRUCTION:
        The scraped text is from the E-commerce page of a website.
        Your job is to extract the Product's infromation and return them in JSON format containing the
        following keys: `Product name`, `Product price`, `Product description(20 words)' link to product.
        Only return in JSON format
        ### VALID JSON (NO PREAMBLE):
        """
)

chain_extract = prompt_extract | llm
res = chain_extract.invoke(input={'page_data':cleaned_text})
type(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```

```python
import pandas as pd
```

```python
df = pd.json_normalize(json_res)  # Flatten if there are nested structures

# Specify the fields you want to include in the CSV
selected_fields = ["Product name", "Product price", "Product description", "link to product"]  # Adjust fields as needed

# Filter the DataFrame to include only the selected fields
filtered_df = df[selected_fields]

# Save the filtered DataFrame to a CSV file
filtered_df.to_csv("filtered_data.csv", index=False)
print("Data saved to filtered_data.csv")
```
